```python
import pandas as pd
```

```python
import re
```

```python
# Carregar base com os dados a serem tratados
df = pd.read_csv("https://raw.githubusercontent.com/Vitor1213/portfolio-dados/refs/heads/main/Projeto%203%20%E2%80%93%20Limpeza%20e%20Padroniza%C3%A7%C3%A3o%20de%20Base%20de%20Dados%20em%20Excel_Python/base-dados.xlsx", sep='\t')
```

```python
df.head()
```

```python
# Padronizando Nome e Cidade
df["Nome"] = df["Nome"].str.strip().str.title().str.replace(" ", " ", regex=True)
df["Cidade"] = df["Cidade"].str.strip().str.title()
```

```python
df.head()
```

```python
# Padronizando Telefones -> formato +55 11 9XXXX-XXXX
def padronizar_telefone(tel):
    numeros = re.sub(r"\D", "", str(tel)) # remove tudo que não for número
    if len(numeros) == 11:
        return f"+55 {numeros[:2]} {numeros[2:7]}-{numeros[7:]}"
    elif len(numeros) == 10: # faltando o 9
        return f"+55 {numeros[:2]} 9{numeros[2:6]}-{numeros[6:]}"
    return None


df["Telefone"] = df["Telefone"].apply(padronizar_telefone)
```

```python
df.head()
```

```python
# Padronizando Datas para YYYY-MM-DD
df["Data_Cadastro"] = pd.to_datetime(df["Data_Cadastro"], errors="coerce", dayfirst=True).dt.strftime("%Y-%m-%d")
```

```python
df.head()
```

```python
# Ajustando E-mails (remover espaços e letras minúsculas)
df["Email"] = df["Email"].str.strip().str.lower()
```

```python
df.head(10)
```

```python
# Removendo e-mails claramente inválidos (sem @ ou .)
df = df[df["Email"].str.contains(r"^[^@\s]+@[^@\s]+\.[^@\s]+$", regex=True, na=False)]
```

```python
df.head(10)
```

```python
# Removendo duplicados (por Nome + Telefone ou Email)
df = df.drop_duplicates(subset=["Nome", "Telefone", "Email"], keep="first")
```

```python
df.head(10)
```

```python
# Salvando base limpa
df.to_csv("clientes_tratados.csv", index=False, encoding="utf-8-sig")
```
